Temperature converstion using python and also know about when temperature generate the heat wave (code:1)

```python
Temperature = float(input("Enter your Temperature in Degree Celsius: "))
Temperature_in_fahrenheit = (Temperature)*9/5 + 32
print ("Your Temperature in fahrenheit is", Temperature_in_fahrenheit)
if Temperature_in_fahrenheit <=113:
   print ("Heat wave is not generate on this temperature")
else:
   print ("Heat wave run on this temperature","please use water more on this temperature")
```
